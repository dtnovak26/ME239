## Problem 2 - Calculate the expectation and variance of a continuous random variable

Take an exponential random variable, see [Problem 2 of Homework 10](lecture10:homework:problem2),
with rate parameter $\lambda$:

$$
T \sim \text{Exp}(\lambda).
$$

The PDF of the exponential is:

$$
p(t) = \lambda e^{-\lambda t},
$$

for $t\ge 0$ and zero otherwise.

+ Prove mathematically that $\mathbf{E}[T] = \lambda^{-1}$. Hint: You need to do the integral $\int_0^\infty tp(t)dt$ using [integration by parts](https://en.wikipedia.org/wiki/Integration_by_parts).

**Answer:** I start by defining the integral
$$E[T] = \int_0^\infty tp(t)dt = E[T] = \int_0^\infty t(\lambda * e^{-\lambda * t} )dt$$

For integration by parts I used: u = t, dv = $\lambda * e^{-\lambda * t}$, du = dt, v = $-e^{-\lambda * t}$. The expanded integral is:

$$\int_0^\infty t\lambda e^{-\lambda t}dt=\left[-te^{-\lambda t}\right]_0^\infty+\int_0^\infty e^{-\lambda t}dt$$

This simplifies to

$$\int_0^\infty e^{-\lambda * t}dt = \frac{1}{\lambda} = \lambda^{-1}$$
<br><br><br><br><br><br><br><br>

+ Prove mathematically that $\mathbf{V}[T] = \lambda^{-2}$. Hint: Use integration by parts to find $\mathbf{E}[T^2]=\int_0^\infty t^2 p(t)dt$ and then use one of the properties of the variance.

**Answer:** We already have $E[T]$, so we must find $E[T^2]$ using integration by parts

$$E[T^2] = \int_0^\infty t^2(\lambda * e^{-\lambda * t} )dt$$

The integration by parts is similar with $u = t^2$, dv = $\lambda * e^{-\lambda * t}$, du = 2tdt, v = $-e^{-\lambda * t}$. The expanded integral is more complex but is equal to:

$$[-t^2e^{-\lambda t}]_0^\infty+\int_0^\infty 2te^{-\lambda t}dt=2\int_0^\infty te^{-\lambda t}dt$$

This simplifies to:

$$2*\int_0^\infty te^{-\lambda t}dt=\frac{1}{\lambda^2}\int_0^\infty se^{-s}ds=\frac{1}{\lambda^2}\cdot 1!=\frac{2}{\lambda^2}=2\lambda^{-2}$$

Which is funny, because after all that work, it just raised the power on the bottom by one which is what we would expect! Nice to go through the process of proving it though.

Now that we have this, we can simply add the two together as showhn in the original statement to prove the question:
$$V[T] = 2\lambda^{-2} - \lambda^{-2} = \lambda^{-2}$$
