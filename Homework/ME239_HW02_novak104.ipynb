## Problem 2 - Numerical Python

Let's make a random array:

````python
import numpy as np
np.random.seed(12345) # We need this so that we get the same numbers
                      # every time we runt the following code
x = np.random.randn(100)
x
````

Find the minimum of ``x``.

````python
# finding the minimum using a for loop
min = x[1]
for i in x:
  if(i < min):
    min = i
# Checking minimum using the min() function
print(min)
print(x.min())

````

Find the maximum of ``x``.

Find the maximum of ``x``.

````python
# Finding the max by just using max() function
print(x.max())
````

Use ``x.shape`` to get the number of elements of ``x`` as an integer.

````python
# number of elements
numElem = x.shape
print(f'Number of elements = {numElem[0]}')
````

Find the sum of all the elements of ``x`` divided by the number of elements of ``x``. This is the average.

````python
# your code here
sum = 0

for i in x:
  sum = sum + i

calcMean = sum / numElem[0]
````

Compare the average you found above with the result of ``x.mean()``. Are they the same?

````python
# your code here
print(f'Value from x.mean(): {x.mean()}')
print(f'Value I calculated:  {calcMean}')

````

Square the elements of ``x`` and store them in a new array called ``x2``.

````python
# squaring each element can be done easily
x2 = pow(x,2)
print(x2)
````

Find the average of ``x2`` and subtract from it the square of average of ``x``.

````python
# Get x2 Average
x2Avg = x2.mean()

# Subtract the square of average of x from the average of x2
finalVal = x2Avg - pow(calcMean, 2)

print(finalVal)
````

Compare the result you found on the cell above with ``x.var()``. Are they the same? (Remark: You just calculated the empirical variance of the "data" ``x`` in two different ways!)

````python
# Comparing values, they are the same! This should line up and be correct.
print(f'My value:  {finalVal}')
print(f'x.var val: {x.var()}')
````
