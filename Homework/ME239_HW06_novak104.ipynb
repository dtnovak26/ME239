## Problem 1 - Loops and conditionals

Consider the following list:

````python
data = [1, 4, 3, 10, 4, 3, 4, 4]
````

+ Write a loop that computes the average of the elements in the list and ``print`` the result using two significant digits.

````python
# loop to get average of values
sum = 0
total = len(data)
for i in data:
  sum += i
# Printing the average value
print(f'{(sum/total):.2f}')
````

+ Write code that finds the number of times the element 4 occurs in the list. Hint: Use a loop and an if-statement.

````python
countedValue = 4 # Specify value to count
count = 0 # Number of occurances

# For loop for checking value
for i in data:
  # Check for value
  if(i == countedValue):
    count += 1

# Print value
print(f'The value {countedValue} occurs {count} times in the list.')
````

+ Write a Python function that takes a list as an argument and returns the number of times a given element (also passed as an argument to the function) appears in the list. Call that function ``find_number_of_occurences(a, elm)``. Make sure you follow best practices when writing the docstring of your function.

````python
def find_number_of_occurences(a, elm):
  """ Finds the number of occurances of a value (elm) in a given list (a) and
      prints the integer result.

  Arguments:
    a          -   the list that will be searched
    elm        -   the value that will be searched for

    Returns: The number of times the value (elm) occurs in the list (a).
  """
  count = 0 # number of times elm occurs in a

  # For loop for checking value
  for i in a:
    # Check for value specified as elm (from user input.)
    if(i == elm):
      count += 1

  #Originally wasnt sure if a print or return was wanted, decided to just return the count
  # (if it was wanted formatted more neatly as a string the below string in the print could be returned instead)
  # print(f'The value {elm} occurs {count} times in the specified list.')

  return count
````

+ Write a Python function that takes a list as an argument and returns the number of elements that are greater than a given element (also passed as an argument to the function). Call that function ``find_number_of_elms_greater_than(a, elm)``. Make sure you follow best practices when writing the docstring of your function.

````python
def find_number_of_elms_greater_than(a, elm):
  """ Finds the number of occurances of a value greater than the user input
      value (elm) in a given list (a) and returns  the integer result.

  Arguments:
    a          -   the list that will be searched
    elm        -   value compared to list values to determine greater than condition.

    Returns: The number of elements that are greater than the given value (elm).
  """
  count = 0 # number of times a value greater than elm occurs in a

  # For loop for checking value
  for i in a:
    # Check for value being greater than elm
    if(i > elm):
      count += 1

  return count
````
